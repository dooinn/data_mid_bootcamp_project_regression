```python
df_modelling = df.copy()
```

```python
# Drop 'id' column
df_modelling = df_modelling.drop(columns=['id'])

# Convert 'date' column to datetime
df_modelling['date'] = pd.to_datetime(df_modelling['date'])

# Extract year from 'date'
df_modelling['year'] = df_modelling['date'].dt.year

# Drop 'date' column
df_modelling = df_modelling.drop(columns=['date'])
```

# 6. Data Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def train_and_test(models, X, Y):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
    model_results = {}
    for model in models:
        model_name = type(model).__name__
        model.fit(X_train, Y_train)
        predictions = model.predict(X_test)

        mse = mean_squared_error(Y_test, predictions)
        mae = mean_absolute_error(Y_test, predictions)
        r2 = r2_score(Y_test, predictions)

        model_results[model_name] = {'R2': r2, 'MAE': mae, 'MSE': mse}
    return pd.DataFrame(model_results).T

# models = [LinearRegression(), KNeighborsRegressor(), MLPRegressor(max_iter=500)]
models = [LinearRegression()]
X = df_modelling.drop(columns=['price'])
Y = df_modelling['price']

results = train_and_test(models, X, Y)
print(results)
```

# Feature Engineering

- **Age of the house**: We have the year when the house was built (yr_built) and the year of the sale (year). The age of the house at the time of sale might be an important factor in determining the price.

- **Renovation age**: Similar to the age of the house, the number of years since the last renovation could also be important. We can calculate this as the difference between the sale year (year) and the renovation year (yr_renovated). If yr_renovated is 0, we can assume the house was never renovated.

- **Total rooms**: The total number of rooms might be more relevant than the number of bedrooms and bathrooms separately. We can calculate this as the sum of bedrooms and bathrooms.

- **Ratio of living space to lot**: The ratio of the living space (sqft_living) to the lot size (sqft_lot) could be an indicator of how spacious a house feels, which might affect the price.

```python
df_modelling['house_age'] = df_modelling['year'] - df_modelling['yr_built']
df_modelling['renovation_age'] = df_modelling['year'].where(df_modelling['yr_renovated']==0, df_modelling['year'] - df_modelling['yr_renovated'])
df_modelling['total_rooms'] = df_modelling['bedrooms'] + df_modelling['bathrooms']
df_modelling['living_lot_ratio'] = df_modelling['sqft_living'] / df_modelling['sqft_lot']
```

```python
X = df_modelling.drop(columns=['price'])
Y = df_modelling['price']

results = train_and_test(models, X, Y)
print(results)
```

# Outlieres Removed

```python
# Calculate the IQR of the 'price' column
Q1 = df_modelling['price'].quantile(0.25)
Q3 = df_modelling['price'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove the outliers
df_modelling = df_modelling[(df_modelling['price'] >= lower_bound) & (df_modelling['price'] <= upper_bound)]

models = [LinearRegression(), KNeighborsRegressor(), MLPRegressor(max_iter=500)]
X = df_modelling.drop(columns=['price'])
Y = df_modelling['price']

results = train_and_test(models, X, Y)
print(results)

```
